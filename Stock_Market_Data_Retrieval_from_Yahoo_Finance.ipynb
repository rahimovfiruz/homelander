```python
import yfinance as yf
import pandas as pd

# Define the list of tickers
tickers = ['MSFT', 'ORCL', 'ADBE', 'PANW', 'VMW',	'SNPS',	'FTNT',	'CRWD',	'PLTR',	'SQ',	'SPLK',	'ZS','VRSN', 'NET',	'FLT', 'AKAM', 'CHKP', 'OKTA', 'IOT', 'GDDY',	'GEN', 'DOX',	'DBX', 'TOST', 'NTNX', 'FFIV',	'CFLT',	'PATH',	'HOOD',	'WEX', 'CCCS', 'CYBR', 'SPSC', 'INFA', 'QLYS', 'AFRM', 'ALTR', 'DLO',	'S', 'TENB', 'RELY', 'WIX', 'TDC', 'FOUR', 'HCP',	'SQSP',	'EEFT',	'BOX', 'ESMT', 'VRNS']

# Create empty lists to store the data
balance_sheet_data = []
income_statement_data = []
cash_flow_data = []

# Fetch the financial data for each ticker
for ticker in tickers:
  stock = yf.Ticker(ticker)

# Fetch balance sheet data
  balance_sheet = stock.balance_sheet.T
  balance_sheet['Ticker'] = ticker
  balance_sheet_data.append(balance_sheet)

# Fetch income statement data
  income_statement = stock.financials.T
  income_statement['Ticker'] = ticker
  income_statement_data.append(income_statement)

# Fetch cash flow data
  cash_flow = stock.cashflow.T
  cash_flow['Ticker'] = ticker
  cash_flow_data.append(cash_flow)

# Concatenate the data into DataFrames
  df_balance_sheets = pd.concat(balance_sheet_data).reset_index()
  df_income_statements = pd.concat(income_statement_data).reset_index()
  df_cash_flows = pd.concat(cash_flow_data).reset_index()

# Display the DataFrames
print("Balance Sheets:")
print(df_balance_sheets)
print("\nIncome Statements:")
print(df_income_statements)
print("\nCash Flows:")
print(df_cash_flows)
```

```python
import yfinance as yf
import pandas as pd

# Define the list of tickers
tickers = ['MSFT',	'ORCL',	'ADBE',	'PANW',	'VMW', 'SNPS', 'FTNT',	'CRWD',	'PLTR',	'SQ',	'SPLK',	'ZS','VRSN',	'NET',	'FLT',	'AKAM', 'CHKP',	'OKTA',	'IOT',	'GDDY',	'GEN', 'DOX',	'DBX', 'TOST', 'NTNX',	'FFIV',	'CFLT',	'PATH', 'HOOD',	'WEX',	'CCCS',	'CYBR',	'SPSC',	'INFA',	'QLYS',	'AFRM',	'ALTR',	'DLO',	'S',	'TENB',	'RELY',	'WIX',	'TDC', 'FOUR',	'HCP',	'SQSP',	'EEFT',	'BOX',	'ESMT',	'VRNS']

# Create empty DataFrames for balance sheet, income statement, and cash flow
df_balance_sheet_r = pd.DataFrame()
df_income_statement_r = pd.DataFrame()
df_cash_flow_r = pd.DataFrame()
# Fetch the quarterly financial data for each ticker for ticker in tickers:
for ticker in tickers:
  stock = yf.Ticker(ticker)
  balance_sheet = stock.balance_sheet
  income_statement = stock.financials
  cash_flow = stock.cashflow
# Reset the index column name to "Account"
  balance_sheet = balance_sheet.rename_axis("Account").reset_index()
  income_statement = income_statement.rename_axis("Account").reset_index()
  cash_flow = cash_flow.rename_axis("Account").reset_index()
# Add the ticker/symbol column
  balance_sheet['Ticker'] = ticker
  income_statement['Ticker'] = ticker
  cash_flow['Ticker'] = ticker
# Append the data to the respective DataFrames
  df_balance_sheet_r = pd.concat([df_balance_sheet_r, balance_sheet], axis=0)
  df_income_statement_r = pd.concat([df_income_statement_r, income_statement], axis=0)
  df_cash_flow_r = pd.concat([df_cash_flow_r, cash_flow], axis=0)

# Display the DataFrames
print("Balance Sheet Data:")
print(df_balance_sheet_r)
print("\nIncome Statement Data:")
print(df_income_statement_r)
print("\nCash Flow Data:")
print(df_cash_flow_r)
```

```python
df_balance_sheets.to_excel("df_balance_sheets.xlsx")

df_income_statements.to_excel("df_income_statements.xlsx")

df_cash_flows.to_excel("df_cash_flows.xlsx")

df_balance_sheet_r.to_excel("df_balance_sheet_r.xlsx")

df_income_statement_r.to_excel("df_income_statement_r.xlsx")

df_cash_flow_r.to_excel("df_cash_flow_r.xlsx")

df_historical_data.to_excel("df_historical_data.xlsx")

df_summary.to_excel("df_summary.xlsx")
```
